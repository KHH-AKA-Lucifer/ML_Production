```python
# import required libraries 
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# get current directory
root_path = os.getcwd()
```

```python
# read the image 
img_path = root_path+r'/test_imgs/shop_invoice.png'
input_img = cv2.imread(img_path)
print("shape: ", input_img.shape)
plt.imshow(input_img)
```

```python
# convert to grayscale image
gray_img = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
threshold_img = cv2.threshold(gray_img, 50, 100, 8)[1]
plt.imshow(threshold_img)
```

```python
canny_edges = cv2.Canny(threshold_img, 50, 220, None, 3)
plt.imshow(canny_edges)
```

```python
minLineLength = 100
maxLineGap = 10
lines = cv2.HoughLinesP(canny_edges, 1, np.pi/180, 10, None, minLineLength, maxLineGap)
print(len(lines))
print(lines[0])
```

```python
start_point = (lines[0][0][0], lines[0][0][1])
print(start_point)
```
